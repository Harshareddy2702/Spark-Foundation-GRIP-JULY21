```python
# Importing all the neccessory modules
import pandas as pd             #It reads the file and create dataframes
import numpy as np              #It calculates the mathematical problems
import matplotlib.pyplot as plt #It is used plotting data from url and trained data
%matplotlib inline
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url)
print("Data Taken in :")
df.head(20)                     # Even required is 9 hours,we will implement it for 20 hours 
```

```python
df.plot(x = "Hours",y = "Scores" ,style= "x")
plt.title("HOURS VS PERCENTAGE")
plt.xlabel("No.of.Hours Studied -->")
plt.ylabel("Percentage obtained -->")
plt.show()
```

```python
# Selecting the values of data from the dataframe
X = df.iloc[:,:-1].values
Y =df.iloc[:,1].values
```

```python
# Splitting the data values obtained into trainiing and testing samples :
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
# Sample data taken for training are trained using Linear Regression model's algorithm
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
print("Trained")
```

```python
# Plotting the trained regreesion model's output using a line equation
line= regressor.coef_*X + regressor.intercept_
plt.scatter(X,Y)
plt.plot(X,line)
plt.show()
```

```python
y_pred = regressor.predict(X_test)
print(X_test)
df1 = pd.DataFrame({'Actual':Y_test, 'predicted': y_pred})
df1
```

```python
# What Will be the student score, if he studies for any given amount of time can be calculated directly using predict function
hours = 9.25
own_pred = round(regressor.predict([[9.25]]) [0],2)
print("No of Hours = {}".format(hours))
print("Predicted Score ={}".format(own_pred))
```

```python
from sklearn import metrics
print('Mean Absolute Error :',metrics.mean_absolute_error(Y_test,y_pred))
```
